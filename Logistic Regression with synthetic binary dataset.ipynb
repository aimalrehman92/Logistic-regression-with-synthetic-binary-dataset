# Logistic Regression with synthetic binary dataset

Objective: Perform binary classification task

```python
import numpy as np
```

```python
def sigmoid(x):
    y = 1/(1+np.exp(-x))
    return y;
```

```python

n_feat = input("Enter the number of features: ")
n_feat = int(n_feat)

n_samples = input("Enter the number of total samples in the dataset: ")
n_samples = int(n_samples)
```

Split the dataset into features and label

```python

X = np.random.randint(2, size=(n_feat+1, n_samples))    				# the whole dataset

X_feat = X[:n_feat,:]													# features in dataset
y = X[n_feat, :]														# label in dataset
```

Split the dataset into train and test part

```python

n_train = int(0.8*n_samples)											# 80% of the samples will be used for training


X_feat_train = X_feat[:, :n_train]
X_feat_test = X_feat[:, n_train:]										# split the dataset into train and test

label_train = X[n_feat, :n_train]
label_test =  X[n_feat, n_train:]										# labels in both training data and in test data

```

In the below cell, we perform training of the weights of linear function of the features

```python

J = 0																	# cost function for logistic regression

W = np.array(np.zeros(n_feat))									 		# weights for the feratures invovled : 11 features
W = W.reshape(len(W), 1) 												# shape = (11,1)

B = 0;							

z = np.dot(X_feat_train.T, W) + B 										# outcomes corresponding to samples

a = sigmoid(z)															# logistic function

dz = a*(1-a)

B += (1./n_train)*np.sum(dz)


W = np.array((1./n_train)*np.dot(X_feat_train, dz))						#already normalized W
```

Now we go and test the trained weights

```python

X_feat_test = X_feat[:, n_train:]										# features in test data
label_test =  X[n_feat, n_train:]										# labels in test data

wrongs = 0
num_to_test_on = int(0.2*n_samples)
for i in range(num_to_test_on):
    case = np.array(X_feat_test.T[i]).reshape(n_feat, 1)
    case_label = label_test[i]
    
    output = sigmoid(np.dot(W.T, case) + B)
    estimated_label = 1 if output>0.5 else 0
    wrongs += estimated_label^case_label
    print ("estimated label:", estimated_label, "  actual label:", case_label)
```

```python
print(wrongs)
```

```python
print('Accuracy:' + str(100*(num_to_test_on-wrongs)/num_to_test_on)+ '%')
```
